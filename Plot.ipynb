```python
import os
import numpy as np
import pandas as pd
import matplotlib
```

```python
# Data Files Locations

file_loc_train = "data_miccai/train"
file_loc_test = "data_miccai/test"
file_loc_val = "data_miccai/val"
```

```python
# list of all the folders in the train location 
list_of_dir  = os.listdir(file_loc_train)

print("Total folders :", len(list_of_dir))
```

```python
risk1 = []
risk2 = []
risk3 = []

for i in list_of_dir:
    #print(file_loc_train+"/"+i)
    
    cd_list_of_dir = file_loc_train+"/"+i
    file_read = pd.read_csv(cd_list_of_dir+"/outcome.csv", sep=",")
    
    val = file_read['risk_1']
    if (val[0]  == True):
        risk1.append(1)
    else:
        risk1.append(0)
    
    val = file_read['risk_2']
    if (val[0]  == True):
        risk2.append(1)
    else:
        risk2.append(0)
    
    val = file_read['risk_3']
    if (val[0]  == True):
        risk3.append(1)
    else:
        risk3.append(0)
```

```python
sum_risk = [0]*8

print(sum_risk)

for i in range(len(risk1)):
    sum_risk[risk1[i]*4 + risk2[i]*2 + risk3[i]]+=1
```

```python
print(sum_risk)
```

```python
sm = 0
for i in sum_risk:
    sm+=i
print(sm)
```

```python
import matplotlib.pyplot as plt

xpoints = np.array([0,1,2,3,4,5,6,7])
ypoints = np.array(sum_risk)
plt.plot(xpoints, ypoints)
plt.show()
```
